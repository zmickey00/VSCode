```python
import numpy as np
allowed_chars=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
allowed_char_dict={char:i for i,char in enumerate(allowed_chars)}
transition_matrix=np.zeros((26,26))
```

```python
with open("warandpeace.txt") as book:
    lines = book.readlines()
for line in lines:
    for punc in ['.', '?', '!', '...', ',', ':', ';','“','”',"‘","’",'(',')']:
        line = line.replace(punc, "").replace('-',' ').replace('—',' ')
    words=line.split()
    for word in words:
        word=word.strip().lower()
        if not any(letter not in allowed_chars for letter in word):
            for i in range(len(word)-1):
                transition_matrix[allowed_char_dict[word[i]],allowed_char_dict[word[i+1]]]+=1
```

```python
for i in range(26):
    transition_matrix[i,:]=transition_matrix[i,:]/np.sum(transition_matrix[i,:])
```

```python
import pandas as pd 
df = pd.DataFrame(transition_matrix)
df
```

```python
df.to_csv("war_and_peace_transition_matrix_letters_only.csv", header=False, index=False)
```
